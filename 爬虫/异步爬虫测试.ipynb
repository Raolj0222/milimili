```python
import asyncio
import requests
import aiohttp 
# aiohttp 是一个用于异步 HTTP 请求的 Python 库，能让你“同时”访问多个网页，而不是一个一个排队访问
from bs4 import BeautifulSoup
```

```python
headers = {
    "User-Agent": "Mozilla/5.0"
}
semaphore = asyncio.Semaphore(20)  # 控制最大并发数量（建议 10~50）
```

```python
# 获取网页详情信息
base_url = "https://phonedb.net"
list_url = base_url + "/index.php?m=device&s=list&filter={}"
def get_phone_info(page_id):
    url = list_url.format(page_id)
    res = requests.get(url,headers=headers,verify=False)
    soup = BeautifulSoup(res.text, "html.parser")
    # 页面详情页数据存放在一个 div 标签中，通过class属性定位标签信息
    info = soup.find_all("div", class_="content_block_title")
    info_dic={}
    title_list = []
    url_list =[]
    for i in info:
        # 标题信息和对应超链接地址存放在a标签中
        row = i.find("a")
        title_list.append(row.get_text())
        url_list.append(base_url+row.get('href','').replace("\n", "").strip())
    info_dic['title'] = title_list
    info_dic['url'] = url_list
    return info_dic
info = get_phone_info('58')
```
