```python
ng = requests.post(url, headers=headers, json=body1)
login =requests.post(url, headers=headers, json=body2)
```

```python
login_df=pd.DataFrame(login.json()['data']['data'])[['ssid','#vp@event_date']]
ng_df=pd.DataFrame(login.json()['data']['data'])[['ssid','#vp@event_date']]
```

```python
ng_df['首次付费时间'] = pd.to_datetime(ng_df['#vp@event_date'],unit='ms')
login_df['活跃时间'] = pd.to_datetime(login_df['#vp@event_date'],unit='ms')
```

```python
res = ng_df.merge(login_df[['ssid','活跃时间']],how='left',on='ssid')
```

```python
res['生命周期']= (res['首次付费时间'] -res['活跃时间']).dt.days
res['付费日期'] =res['首次付费时间'].dt.strftime('%Y-%m-%d')
```

```python
for i in range(0,42):
    print(i)
    res[f'{i}-D留存'] = res.apply(lambda x: x['ssid'] if x['生命周期']==i else np.nan,axis=1 )
    
```

```python
res.pivot_table(values="ssid",# 计算字段
                      index="付费日期",# 分组的行
                      columns="生命周期",# 分组的列
                      aggfunc="nunique", # 聚合方式
                      fill_value=0, # 对缺失值的填充，默认为nan
                      margins=True, # 是否启用总计
                      margins_name='总计' # 总计的名称
                     )
```
