```python
import pandas as pd
```

```python
data = pd.read_csv('../临时文件/进入扭蛋用户.csv',index_col=False)
```

```python
df=data[['用户App登录ID','事件发生时间','生命周期天数']]
```

```python
df['时间']=pd.to_datetime(df['事件发生时间'],unit='ms')
```

```python
# 根据用户时间
df['顺序'] = df.groupby(by=['用户App登录ID'])['时间'].rank(ascending=True,method='first')
```

```python
res = df.query('顺序==1')
```

```python
res['生命周期天数'].median()
```

```python
res['生命周期天数'].mode()
```
