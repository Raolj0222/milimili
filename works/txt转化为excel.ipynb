```python
import pandas as pd
```

```python


# 1. 设置文件路径（改成你自己的 txt 文件路径）
txt_file = r"D:\Desktop\iOS收入明细\7月数据\FD_92395812_0725.txt."
excel_file = r"D:\Desktop\iOS收入明细\7月数据\7月销售数据.xlsx"

# 2. 读取 TXT 内容
with open(txt_file, 'r', encoding='utf-8') as file:
    lines = file.readlines()

# 3. 跳过元信息，提取表头和数据
start_idx = 3  # 从第4行开始是表头
header = lines[start_idx].strip().split('\t')
data_lines = lines[start_idx + 1:]

# 4. 解析数据为列表
data = [line.strip().split('\t') for line in data_lines if line.strip()]

# 5. 创建 DataFrame 并保存为 Excel
df = pd.DataFrame(data, columns=header)
df.to_excel(excel_file, index=False)

print(f"转换完成，Excel 文件已保存为：{excel_file}")
```
