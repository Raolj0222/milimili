```python
res = ng_df.merge(login_df[['ssid','活跃时间']],how='left',on='ssid')
```

```python
ng_df.groupby(by='付费日期',as_index=False)['ssid'].nunique()
```

```python
for i in range(0,42):
    print(i)
    res[f'{i}-D留存'] = res.apply(lambda x: x['ssid'] if x['生命周期']==i else np.nan,axis=1 )
    
```

```python
col_list=['ssid']+[f'{i}-D留存'  for i in range(42)]
df=res.groupby(by=['付费日期'],as_index=False)[col_list].agg({
    f'{j}':'nunique' for j in col_list
})
```
