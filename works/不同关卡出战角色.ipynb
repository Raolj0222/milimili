```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('../临时文件/5-30关免费用户闯关数据.csv',index_col=False)[['level_name','level_end_info','是否通关']]
```

```python
role_list =[str(i) for i in range(101,127)]
```

```python
for i in role_list:
    data[f'{i}00'] = data['level_end_info'].astype('str').str.count(i)
```

```python
role_col = [str(i)+'00' for i in range(101,127)]
```

```python
# data['level_name'] = '22关'
```

```python
res=data.query('是否通关==1').groupby(by='level_name',as_index=False)[['level_end_info']+role_col].agg({'level_end_info':'count'}|{i:'sum' for i in role_col})
```

```python
res.to_excel('../临时文件/5-30关免费用户通关出战角色.xlsx',index=False)
```
