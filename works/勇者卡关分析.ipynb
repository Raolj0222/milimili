```python
import pandas as pd
import numpy as np
```

```python
# 闯关数据 
data1=pd.read_csv('../临时文件/用户闯关数据.csv',index_col=False)[['Metric','用户唯一id','total_pay_num','事件发生时间']]
```

```python
data1['ssid'].value_counts()
```

```python
data1['Metric']='闯关'
```

```python
data1.columns=['event','ssid','累计付费次数','时间']
```

```python
data1['时间']=pd.to_datetime(data1['时间'], unit='ms')
```

```python
data1.head()
```

```python
# 付费数据
data2=pd.read_csv('../临时文件/用户付费数据.csv',index_col=False)[['Metric','用户唯一id','事件发生时间','Total']]
```

```python
data2['Metric']='内购'
```

```python
data2.columns=['event','ssid','时间','付费金额']
```

```python
data2['时间']=pd.to_datetime(data2['时间'], unit='ms')
```

```python
data2.head()
```

```python
# 合并数据
res=pd.concat((data1,data2))
```

```python
res['ssid'].value_counts()
```

```python
res.to_excel('闯关付费数据汇总.xlsx')
```

```python
# 根据时间排序
res.sort_values(by=['ssid','时间'],ascending=[True,True],inplace=True)
```

```python
res.head()
```

```python
# 聚合事件
df=res.groupby(by=['ssid'],as_index=False)['event'].agg(lambda x:','.join(x))
```

```python
df['ssid'].value_counts()
```

```python
df['event_list']=df['event'].str.split(',')
```

```python
df.head()
```

```python
# 计算闯关失败次数 --> 付费路径
def get_level_ng_info(x):
    res_list=[]
    tmp_event = x[0]
    tmp_event_num=1
    if len(x)==1:
        res_list.append({tmp_event:tmp_event_num})
    else:
        for i in x[1:]:
            if i!=tmp_event:
                res_list.append({tmp_event:tmp_event_num})
                tmp_event=i
                tmp_event_num=1
                
            else:
                tmp_event_num +=1
        res_list.append({tmp_event:tmp_event_num})
    return res_list
```

```python
df['失败-付费路径']=df.apply(lambda x:get_level_ng_info(x['event_list']),axis=1)
```

```python
df.to_excel('../临时文件/用户闯关-付费路径.xlsx')
```
