```python
import pandas as pd
import numpy as np
```

```python
df1=pd.read_csv('../临时文件/关卡付费.csv',index_col=False)
```

```python
df2=pd.read_csv('../临时文件/关卡活跃.csv',index_col=False)
```

```python
df1 = df1[['当前最大章节id','Total']]
```

```python
df2 = df2[['level_id','Total']]
```

```python
df1.columns=['关卡id','付费金额']
df2.columns=['关卡id','活跃人数']
```

```python
res=df2.merge(df1,how='left',on='关卡id')
```

```python
res=res.sort_values(by=['关卡id'],ascending=[True])
```

```python
res1=res.dropna(axis=0,how='any')
```

```python
res1['累计付费']=res1['付费金额'].cumsum()
```

```python
res1['初始用户数'] = 12875 
```

```python
res1['用户ltv']=res1['累计付费']/res1['初始用户数']
```

```python
res1.to_excel('../临时文件/关卡ltv变化.xlsx',index=False)
```
