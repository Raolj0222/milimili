```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../临时文件/9-28-10-12用户付费.csv',index_col=False)[['用户App登录ID','首次安装App时间','事件发生时间','Total']]
```

```python
df['install_time']= pd.to_datetime(df['首次安装App时间'])
df['ng_time']=pd.to_datetime(df['事件发生时间'],unit='ms')
```

```python
# 计算内购时生命周期天数
df['live_day'] = (df['ng_time'] - df['install_time']).dt.days
```

```python
# 分组
df['cuts'] = pd.cut(
    df['live_day'],# 需要分组的列
    bins=[0,7,14,21,1000],# 分组的组数或者自定义分组列表
    labels=['第一周','第二周','第三周','第四周'],
    right=False,# 是否包含右边的值
    include_lowest=True # 是否包含最低值
    
)
```

```python
df['第一周付费用户'] = df.apply(lambda x:1 if x['cuts']=='第一周' else 0 ,axis=1)
df['前二周付费用户'] = df.apply(lambda x:1 if x['cuts'] in ['第一周','第二周'] else 0 ,axis=1)
df['前三周付费用户'] = df.apply(lambda x:1 if x['cuts']in ['第一周','第二周','第三周']  else 0 ,axis=1)
df['前四周付费用户'] = df.apply(lambda x:1 if x['cuts']in ['第一周','第二周','第三周','第四周']  else 0 ,axis=1)
```

```python
# 不同分组的用户金额分布
df1=df.query('第一周付费用户==1').groupby(by='用户App登录ID',as_index=False)['Total'].sum()
df1['区间']=pd.cut(
    df1['Total'],
    bins=[0,20,40,60,80,100,100000],# 分组的组数或者自定义分组列表
    labels=['0-20','20-40','40-60','60-80','80-100','100+'],
    right=False,# 是否包含右边的值
    include_lowest=True # 是否包含最低值
)
# 不同区间的人数
df1=df1.groupby(by='区间',as_index=False)[['用户App登录ID','Total']].agg({
    '用户App登录ID':'nunique',
    'Total':'sum'
})
```
