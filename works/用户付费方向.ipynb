```python
# 钻石卡池
sql3 = """
select "#account_id","#event_time",
case when "$part_date" between '2023-12-06' and '2023-12-24' then '奏'
     when "$part_date" between '2023-12-25' and '2024-01-14' then '卡莎' 
     when "$part_date" between '2024-01-15' and '2024-02-12' then '蕾娜'
     when "$part_date" between '2024-02-13' and '2024-03-11' then '银蛇'
end as "role_name"
from  
ta.v_event_33 
where "$part_event" = 'Activity_recruit'
and "$part_date" between '2023-12-01' and '2024-03-11'
and "type" = '23'
"""

# 盲盒卡池
sql4= """
select "#account_id","#event_time",
case when "$part_date" between '2023-12-11' and '2023-12-31' then '维帕尔'
     when "$part_date" between '2024-01-01' and '2024-01-28' then '月' 
     when "$part_date" between '2024-01-29' and '2024-02-25' then '桃子'
     when "$part_date" between '2024-02-26' and '2024-03-11' then '莲'
end as "role_name"
from  
ta.v_event_33 
where "$part_event" = 'Activity_recruit'
and "$part_date" between '2023-12-01' and '2024-03-11'
and "type" = '22'
"""
```

```python
role_get_zuanshi = get_sqldata(token,sql3)
```

```python
role_get_zuanshi.columns=['玩家id','抽卡时间','卡池角色']
```

```python
role_get_manghe = get_sqldata(token,sql4)
```

```python
role_get_manghe.columns=['玩家id','抽卡时间','卡池角色']
```

```python
# 连续参与漏斗
role_get_zuanshi['奏']=role_get_zuanshi.apply(lambda x: 1 if x['卡池角色']=='奏' else 0,axis=1)
role_get_zuanshi['卡莎']=role_get_zuanshi.apply(lambda x: 1 if x['卡池角色']=='卡莎' else 0,axis=1)
role_get_zuanshi['蕾娜']=role_get_zuanshi.apply(lambda x: 1 if x['卡池角色']=='蕾娜' else 0,axis=1)
role_get_zuanshi['银蛇']=role_get_zuanshi.apply(lambda x: 1 if x['卡池角色']=='银蛇' else 0,axis=1)
```

```python
# 抽卡参与人数
role_get_zuanshi.groupby(by='卡池角色',as_index=False)['玩家id'].nunique()
```

```python
role_get_manghe['维帕尔']=role_get_manghe.apply(lambda x: 1 if x['卡池角色']=='维帕尔' else 0,axis=1)
role_get_manghe['月']=role_get_manghe.apply(lambda x: 1 if x['卡池角色']=='月' else 0,axis=1)
role_get_manghe['桃子']=role_get_manghe.apply(lambda x: 1 if x['卡池角色']=='桃子' else 0,axis=1)
role_get_manghe['莲']=role_get_manghe.apply(lambda x: 1 if x['卡池角色']=='莲' else 0,axis=1)
```

```python
role_get_manghe.groupby(by='卡池角色',as_index=False)['玩家id'].nunique()
```
