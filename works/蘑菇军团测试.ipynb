```python
import requests 
import pandas 
```

```python
def process(n):
    # 浇水记录、产生孢子和产生道具
    res=[1,0,0,0,0,0]
    # 获取token
    url_token ='http://192.168.0.166:8080/api/v1/userToken'
    body_token={'provider':'TELEGRAM',
           'debug':10000+n,
            'user_data':'rao'
           }
    herders={'Content-Type': 'application/x-www-form-urlencoded',}
    token_data = requests.post(url_token,body_token,herders)
    token=token_data.json()['model']['token']
    header = {
        'Content-Type': 'application/x-www-form-urlencoded',
        'token':token
    }
    # 模拟浇水
    url1 ='http://192.168.0.166:8080/api/v1/doOnceWatering'
    body1={
    'property_id':10000
    }
    data1  = requests.post(url=url1,headers=header,data=body1)
    from_ids = [item['from_id'] for item in data1.json()['model']['benefits']]
    from_types = [item['type'] for item in data1.json()['model']['benefits']]
    
    if  'mushroom_spore_output' in from_types:
        res[1]=1
        # 模拟开孢子 
        url2='http://192.168.0.166:8080/api/v1/pokeSpore'
        body2 = {
            'property_id':from_ids[0],
            'open_number':1  
        }
        data2 = requests.post(url=url2,headers=header,data=body2)
        # 记录产生的水滴
        # print(data2.json())
        typelist = [item['type'] for item in data2.json()['model']['benefits']]
        numberlist  = [item['number'] for item in data2.json()['model']['benefits']]
        if 'spore_water_output' in typelist:
            # num1 = numberlist[typelist.index('spore_water_output')]
            res[2] = 1
        else:
            pass
        # 记录产生的碎片
        if 'spore_mushroom_output' in typelist:
            # num2 = numberlist[typelist.index('spore_mushroom_output')]
            res[3]=1
        else:
            pass
        # 记录产生的木屑
        if 'spore_sawdust_output' in typelist:
            # num3 = numberlist[typelist.index('spore_sawdust_output')]
            res[4]=1
        else:
            pass
        # 记录产生的代币
        if 'spore_coin_output' in typelist:
            # num3 = numberlist[typelist.index('spore_sawdust_output')]
            res[5]=1
        else:
            pass
    
    else:
        pass
    return res  
```

```python
res4=[]
# 开启测试，总共1000000滴水
for i in range(9000,10000):
    # 每次开100滴水
    df_list=[0,0,0,0,0,0]
    for j in range(0,100):
        df=process(i)
        df_list[0] = df[0]+df_list[0]
        df_list[1] = df[1]+df_list[1]
        df_list[2] = df[2]+df_list[2]
        df_list[3] = df[3]+df_list[3]
        df_list[4] = df[4]+df_list[4]
        df_list[5] = df[5]+df_list[5]
    res4.append(df_list)
```

```python
df4= pandas.DataFrame(res4)
df4.columns=['浇水数','产生孢子数','爆出水滴数','爆出碎片数','爆出木屑数','爆出代币数']
df4.to_excel('蘑菇军团测试0918.xlsx')
```
