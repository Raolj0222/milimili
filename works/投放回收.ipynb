```python
# 计算时间差
res1['ng_days']=(res1['内购时间'] -res1['安装时间']).dt.days
res2['ad_days']=(res2['广告时间'] -res2['安装时间']).dt.days
```

```python
res1['ng_cuts'] = pd.cut(
    res1['ng_days'],# 需要分组的列
    bins=[0,7,14,21,28,35,42,100],# 分组的组数或者自定义分组列表
    right=False,# 是否包含右边的值
    labels=[f'第{i}周内购' for i in range(1,8)],# 显示标签
    include_lowest=True # 是否包含最低值
)
res2['ad_cuts'] = pd.cut(
    res2['ad_days'],# 需要分组的列
    bins=[0,7,14,21,28,35,42,100],# 分组的组数或者自定义分组列表
    right=False,# 是否包含右边的值
    labels=[f'第{i}周广告' for i in range(1,8)],# 显示标签
    include_lowest=True # 是否包含最低值
)
```

```python
res_install_ng_ad['7_day_group'] = (res_install_ng_ad['安装时间'] - res_install_ng_ad['安装时间'].min())//timedelta(days=7)+1
```

```python
def custom_weekly_group(row):
    if row["安装时间"] <= pd.Timestamp("2024-09-29"):
        return "2024-09-28~2024-09-29"
    else:
        # Calculate the start of each 7-day group beginning from September 30
        start_date = pd.Timestamp("2024-09-30") + timedelta(days=((row["安装时间"] - pd.Timestamp("2024-09-30")).days // 7) * 7)
        end_date = start_date + timedelta(days=6)
        return f"{start_date.strftime('%Y-%m-%d')}~{end_date.strftime('%Y-%m-%d')}"
res_install_ng_ad["7_day_group"] = res_install_ng_ad.apply(custom_weekly_group, axis=1)
```
